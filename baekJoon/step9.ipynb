```python
#10872
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
    
print(factorial(int(input())))
```

```python
#10870
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(int(input())))
```

```python
def hanoi(n, a1, a2, a3):
    if a3 == n:
        print("")
    elif a1 > a3:
        print(1,3)
        return hanoi(n, a1-1, a2, a3+1)
    elif a1 > a2:
        print(1,2)
        return hanoi(n, a1-1, a2+1, a3)
    elif a2 > a1:
        print(2,1)
        return hanoi(n, a1+1, a2-1, a3)
    elif a2 > a3:
        print(2,3)
        return hanoi(n, a1, a2-1, a3+1)
    elif a3 > a1:
        print(3,1)
        return hanoi(n, a1+1, a2, a3-1)
    else:
        print(3,2)
        return hanoi(n, a1, a2+1, a3-1)
        

n = int(input())
#a1  1, a2  2
a1 = n
a2 = 0
a3 = 0
print(hanoi(n, a1, a2, a3))
```

```python
#2447 from https://lucian-blog.tistory.com/57 
def draw_stars(n):
  if n==1:
    return ['*']

  Stars=draw_stars(n//3)
  L=[]

  for star in Stars:
    L.append(star*3)
  for star in Stars:
    L.append(star+' '*(n//3)+star)
  for star in Stars:
    L.append(star*3)

  return L

N=int(input())
print('\n'.join(draw_stars(N)))
```
